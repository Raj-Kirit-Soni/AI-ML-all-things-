```python
# ==========================================
# Operator Precedence and Associativity
# ==========================================

print("=== Operator Precedence ===")

# Multiplication has higher precedence than addition
result1 = 10 + 5 * 2

print("10 + 5 * 2 =", result1)

# Using brackets changes precedence
result2 = (10 + 5) * 2

print("(10 + 5) * 2 =", result2)
```

```python
# ==========================================

print("\n=== Associativity ===")

# Left to Right Associativity
result3 = 20 - 5 - 3

print("20 - 5 - 3 =", result3)

# Explanation:
# (20 - 5) - 3 = 12
```

```python
# ==========================================

# Right to Left Associativity
result4 = 2 ** 3 ** 2

print("2 ** 3 ** 2 =", result4)

# Explanation:
# 2 ** (3 ** 2)
# 2 ** 9 = 512
```

```python
# ==========================================

print("\n=== Mixed Example ===")

value = 10 + 2 * 3 ** 2

print("10 + 2 * 3 ** 2 =", value)

# Step-by-step:
# 3 ** 2 = 9
# 2 * 9 = 18
# 10 + 18 = 28

# ==========================================

print("\nProgram Finished Successfully")
```
