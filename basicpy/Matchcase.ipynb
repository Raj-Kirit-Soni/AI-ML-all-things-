##Match-case



```python
http_status = int(input("Enter the http code"))

match http_status :
    case 200:
        print(" the server is Running properly")
    case 404:
        print(" Bad Request")
    case 500:
        print("Internal server error")
    case _:
        print("Enter code is invalid")
```

```python
color = input("Enter the color name ")

match color:
    case "red":
        print(" Vehicle STOP")
    case "green":
        print(" Vehicle GO")
    case "yelow":
        print("Vehicle slow")
    case _:
        print("Enter color is invalid")
    
```
