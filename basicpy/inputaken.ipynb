```python
# =====================================
# TAKING INPUT IN DIFFERENT WAYS
# =====================================

# 1. String Input
name = input("Enter your name: ")
print("Name:", name)
```

```python
# =====================================

# 2. Integer Input
age = int(input("\nEnter your age: "))
print("Age:", age)
```

```python
# =====================================

# 3. Float Input
height = float(input("\nEnter your height: "))
print("Height:", height)
```

```python
# =====================================

# 4. Multiple Inputs in One Line
a, b = input("\nEnter two numbers: ").split()

print("First Number:", a)
print("Second Number:", b)
```

```python
# =====================================

# 5. Multiple Integer Inputs
x, y = map(int, input("\nEnter two integers: ").split())

print("Addition:", x + y)
```

```python
# =====================================

# 6. Multiple Float Inputs
p, q = map(float, input("\nEnter two float values: ").split())

print("Multiplication:", p * q)
```

```python
# =====================================

# 7. List Input
numbers = list(map(int, input("\nEnter list elements: ").split()))

print("List:", numbers)
```

```python
# =====================================

# 8. Tuple Input
data = tuple(map(int, input("\nEnter tuple elements: ").split()))

print("Tuple:", data)
```

```python
# =====================================

# 9. Boolean Input
is_student = input("\nAre you a student? (True/False): ")

# Convert string to boolean
is_student = is_student == "True"

print("Student Status:", is_student)
```

```python
# =====================================

# 10. Character Input
char = input("\nEnter a character: ")[0]

print("Character:", char)

# =====================================

print("\nProgram Finished Successfully")
```
