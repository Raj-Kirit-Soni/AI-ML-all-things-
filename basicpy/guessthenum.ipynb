```python
def number_guessing_game():
    secret_number = 15  # You can set any secret number here
    
    while True:
        guess = int(input("Guess the number: "))
        
        if guess > secret_number:
            print("Your guess is above the number.")
        elif guess < secret_number:
            print("Your guess is below the number.")
        else:
            print("Congratulations! Your guess matches the secret number.")
            break  # Exit the loop once guessed correctly


# Run the game
number_guessing_game()
```
